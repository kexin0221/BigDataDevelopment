## Exercise 1: Plotting an Analytical Function

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-50, 50, 100)
y = np.power(x, 2)
```

```python
plt.plot(x, y)
```

```python
y_hat = np.power(x, 3)
plt.plot(x, y_hat)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
