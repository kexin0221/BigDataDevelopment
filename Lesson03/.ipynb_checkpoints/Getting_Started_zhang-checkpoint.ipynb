```python
import pyspark
from pyspark import SparkConf
from pyspark.sql import SparkSession
from pyspark.sql import functions as fun
from pyspark.sql.types import *
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
packages = ','.join([
    "com.johnsnowlabs.nlp:spark-nlp_2.4:3.1.2",
])

spark_conf = SparkConf()
spark_conf = spark_conf.setAppName('spark-nlp-book-plc1')
spark_conf = spark_conf.setAppName('master[*]')
spark_conf = spark_conf.set("spark.jars.packages", packages)

spark = SparkSession.builder.config(conf=spark_conf).getOrCreate()
```

```python
texts = spark.sparkContext.wholeTextFiles("hdfs://localhost:9000/data/mini_newsgroups/*")
```

```python
schema = StructType([
    StructField('filename', StringType()),
    StructField('text', StringType()),
])

texts_df = spark.createDataFrame(texts, schema)
```

```python
texts_df.printSchema()
```

```python
texts_df.show()
```

```python
texts_df.show(n=5, truncate=100, vertical=True)
```

```python
texts_df.limit(5).toPandas()
```

```python
texts_df = texts_df.withColumn(
    'newsgroup', 
    fun.split('filename', '/').getItem(5)
)
```

```python
texts_df.limit(5).toPandas()
```

```python
newsgroup_counts = texts_df.groupBy('newsgroup').count().toPandas()
```

```python
newsgroup_counts.plot(kind='bar', figsize=(10, 5))
plt.xticks(
    ticks=range(len(newsgroup_counts)), 
    labels=newsgroup_counts['newsgroup']
)
plt.tight_layout()
plt.savefig('newsgroup.jpg')
plt.show()
```

```python
import sparknlp
from sparknlp.base import *
from sparknlp.annotator import *
from sparknlp.pretrained import PretrainedPipeline
```

```python
spark = sparknlp.start()
print("Spark NLP version: ", sparknlp.version())
print("Apache Spark version: ", spark.version)
## must make sure that the version of spark-nlp is compatible with that of pretrainedpipeline
## the below type of methods are OK!
## designed by zhangzw
pipeline = PretrainedPipeline('explain_document_ml',lang='en') 
# pipeline = PretrainedPipeline('explain_document_dl',lang='en')
# pipeline = PretrainedPipeline.from_disk("hdfs://localhost:9000/offline_pipeline/explain_document_dl_en_2.4.3_2.4_1584626657780/")
```

```python
result = pipeline.annotate('Hellu wrold!')
print(result)
```

```python
texts_df.printSchema()
```

```python
procd_texts_df = pipeline.annotate(texts_df, 'text')
```

```python
procd_texts_df.printSchema()
```

```python
procd_texts_df.show(n=2)
```

```python
procd_texts_df.show(n=2, truncate=100, vertical=True)
```

```python
from sparknlp import Finisher
finisher = Finisher()
finisher = finisher
# taking the lemma column
finisher = finisher.setInputCols(['lemmas'])
# seperating lemmas by a single space
finisher = finisher.setAnnotationSplitSymbol(' ')
finished_texts_df = finisher.transform(procd_texts_df)
finished_texts_df.show(n=1, truncate=100, vertical=True)
```
