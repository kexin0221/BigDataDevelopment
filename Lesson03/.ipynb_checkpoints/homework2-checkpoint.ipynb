```python
# pride-and-prejudice
import findspark
findspark.init()

import pyspark
from pyspark.sql import SparkSession
spark = SparkSession.builder.getOrCreate()
```

```python
texts = spark.sparkContext.wholeTextFiles("hdfs://localhost:9000/data/pride-and-prejudice/*")
```

```python
from pyspark.sql import functions as fun
from pyspark.sql.types import *

schema = StructType([
    StructField('filename', StringType()),
    StructField('text', StringType()),
])

texts_df = spark.createDataFrame(texts, schema)
```

```python
texts_df.printSchema()
```

```python
texts_df.show()
```

```python
texts_df1 = texts_df.drop('filename')
```

```python
texts_df1 = texts_df1.withColumn(
    'chapter_num', 
    fun.split('text', '\n\n').getItem(2)
)
```

```python
texts_df1.show()
```

```python
texts_df1.count()
```

```python
texts_df1.limit(5).toPandas()
```

```python
rdd_df = spark.read.text(\
    "file:///D:/Big-Data-Science/Big-Data-Analysis-with-Python-master/Lesson03/data/english_words.txt").rdd
lines = rdd_df.map(lambda line: line[0])
words_list = lines.collect()
words_list[0:5]
```

```python
from operator import add
import collections
count_words = []
filenames = texts_df.toPandas()['filename']
for chapter in filenames: 
    rdd_df = spark.read.text(chapter).rdd
    lines = rdd_df.map(lambda line:line[0])
    lines.collect()
    splits = lines.flatMap(lambda x:x.split(' '))
    tokens = splits.filter(lambda x:x and x in words_list)
    token_list = tokens.collect()
    count = len(token_list)
    count_words.append(count)
count_words[0:5]
```

```python
len(count_words)
```

```python
from pyspark.sql import Row
row_item = spark.sparkContext.parallelize(count_words)
new_row = Row("word_count")
row_item_todf = row_item.map(new_row).toDF()
```

```python
texts_df1 = texts_df1.toPandas()
row_item_todf = row_item_todf.toPandas()
text_count_df = texts_df1.join(row_item_todf)
text_count_df.head(5)
```

```python
text_count_df.count()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
ax = sns.barplot(x='word_count',y='chapter_num',data=text_count_df)
plt.title("Word count of Chapters")
ax.figure.set_size_inches(10,10)
```

```python
rdd_df = spark.read.text(\
     "file:///D:/Big-Data-Science/Big-Data-Analysis-with-Python-master/Lesson03/data/Stopwordlist.txt").rdd
lines = rdd_df.map(lambda line: line[0])
stop_list = lines.collect()[1:-1]
stop_list[0:5]
```

```python
rdd_df = spark.read.text("hdfs://localhost:9000/data/pride-and-prejudice/*").rdd
lines = rdd_df.map(lambda line:line[0])
lines.collect()
splits = lines.flatMap(lambda x:x.split(' '))
tokens = splits.filter(lambda x:x and x in words_list and x.lower() not in stop_list)
token_list = tokens.map(lambda x:[x,1])
count = token_list.reduceByKey(add).sortBy(lambda x:x[1],ascending=False)
words_count = count.collect()
words_count[0:5]
```

```python
rdd_df = spark.read.text("hdfs://localhost:9000/data/pride-and-prejudice/44.txt").rdd
lines = rdd_df.map(lambda line:line[0])
lines.collect()
splits = lines.flatMap(lambda x:x.split(' '))
tokens = splits.filter(lambda x:x and x in words_list and x.lower() not in stop_list)
token_list = tokens.map(lambda x:[x,1])
count = token_list.reduceByKey(add).sortBy(lambda x:x[1],ascending=False)
words_count = count.collect()
words_count[0:5]
```

```python
x = []
y = []
for key, value in words_count:
    x.append(key)
    y.append(value)
plt.barh(x[:20], y[:20])
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Chapter 44')
plt.show()
```
