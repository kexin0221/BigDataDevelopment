```python
#第一小题
from pyspark import SparkContext
sc=SparkContext()
from pyspark.sql import SQLContext
sqlc=SQLContext(sc)
```

```python
df=sqlc.read.format('com.databricks.spark.csv').options(header = 'true',inferschema = 'true').load('file:///D:/桌面/data/Log_Reg_dataset.csv')
df.show(10)
```

```python
#第二小题
df.printSchema()
```

```python
df.describe('Age','Web_pages_viewed').show()
```

```python
#第三小题
df.groupby('Country').agg({'Age':'mean'}).show(10)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
data=df.toPandas()
sns.boxplot(x='Country',y='Age',data=data)
plt.show()
```

```python
#第四小题
df_4=df.select('Country','Platform','Status')
df_4=df_4.filter(df_4.Status==1)
df_4=df_4.dropDuplicates()
df_4.show()
```

```python
df_4.write.parquet("output_file1.parquet")
```
